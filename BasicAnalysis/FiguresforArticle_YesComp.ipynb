```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

plt.style.use('seaborn-colorblind')

## Read in Excel files of data and question labels
XLS = pd.ExcelFile('../../ComputationSurveyData/LabeledComputationDataColumnRenamed.xlsx')
XLS2 = pd.ExcelFile('../../ComputationSurveyData/Questions.xlsx')

## Parse the appropriate sheets into dataframes
df = XLS.parse('LabeledComputationData')
questions = XLS2.parse('Sheet1')

#outDir = "/Users/caballero/Repos/research/papers/2017_CompSurvey/figs/" ## Work
outDir = "/Users/danny/Repos/research/papers/2017_CompSurvey/figs/"  ## Home

bboxprops = {'facecolor':'white', 'pad':0.5, 'boxstyle':'round', 'edgecolor':'lightgrey', 'alpha':0.7}
```

```python
onlyYesDF = df.loc[df['Q1'] == 1]
```

```python
## Comp on Exams?
count0 = 0
count1 = 0

countedSchools = []

for index, row in onlyYesDF.iterrows():
    
    if(row['Q34']==1):
        count1 += 1
        countedSchools.append(row['NAME'])
    if(row['Q34'] == 0):
        count0 += 1

Schools = len(set(countedSchools))
totalSchools = len(set(onlyYesDF['NAME']))

print('Total number of unique schools: ', totalSchools)
print('Number of schools with at using Exams: ', Schools)

percentSchools = Schools/totalSchools*100
                   
fig8 = plt.figure(num=None, figsize=(4, 3), dpi=300, facecolor='w', edgecolor='k')
ax8 = fig8.add_subplot(111)

plt.bar([0],[percentSchools])
plt.bar([1],[100-percentSchools])

#plt.ylabel('% of Departments')
plt.yticks([])
YesString = 'Yes\n(N=' + str(Schools) + ')'
NoString = 'No\n(N=' + str(totalSchools-Schools) + ')'
plt.xticks([0,1],[YesString,NoString])

plt.axis([-.5,1.5,0,100])
#plt.title('Does anyone in your department\nteach computing in\nadvanced physics (elective)?')

Nstring = str(totalSchools)
ax8.text(1, 85, 'Comp. on Exams', bbox = bboxprops,ha='center')
ax8.text(-0.4,90,'(d)', ha='center')

plt.tight_layout()
plt.savefig(outDir+'CompExams.pdf')
```
