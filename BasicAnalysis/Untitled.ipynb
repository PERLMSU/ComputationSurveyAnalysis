```python
ranks = ['Full Professor','Associate Professor',
         'Assistant Professor','Instructor/Lecturer',
         'Adjunct faculty','Visiting Professor',
         'Research Professor']
indices = np.arange(0,2*len(ranks),2)

plt.figure(figsize = (6,4), dpi = 150)

for rank, index in zip(ranks, indices):
    
    k = df[df['Q166'] == rank]['Q1'].value_counts()
    
    plt.bar(index, k[1]/(k[0]+k[1]), color='C0', width=0.7)
    plt.bar(index+1, k[0]/(k[0]+k[1]), color='C1', width=0.7)
    
    if index < 2*(len(ranks)-1):
        
        plt.plot([index+1.5,index+1.5],[0,1],'k--', alpha = 0.7)

plt.axis([-1, 2*len(ranks), 0, 1])

plt.xticks(np.arange(0.5,2*len(ranks),2), ranks, rotation=90);
plt.xlabel('Faculty Rank')
plt.ylabel('Fraction of\nRespondents ')

plt.legend(handles=[yes_patch,no_patch])

plt.tight_layout()
```
